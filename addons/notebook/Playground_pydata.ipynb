```python
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine 
import pandas as pd
from pandas.io import sql
from pandas.io.sql import read_sql
import datetime as dt
import numpy as np
```

```python
query = """\
SELECT 
    date(starting_date) as date, 
    count(*) as count
FROM events
GROUP BY 1"""
df = read_sql(query, engine)
```
